#1. What is the difference between descriptive statistics and inferential statistics? Explain with examples

Statistics is the branch of mathematics concerned with collecting, organizing, analyzing, and interpreting data. It is broadly divided into descriptive statistics and inferential statistics.

Descriptive Statistics
  Descriptive statistics deals with summarizing and describing the main features of a dataset. It does not make conclusions beyond the data itself.

Purpose:
  To organize data
  To present data in a meaningful way
  To understand what the data shows

Common tools:
   Mean, median, mode
   Range, variance, standard deviation
  Tables, charts, graphs (bar chart, histogram, pie chart)

Example:
  Calculating the average marks of a class
  Finding the highest and lowest salary in a company
  Creating a bar graph of monthly sales


Inferential Statistics
   Inferential statistics uses sample data to make predictions or conclusions about a population.  

Purpose:
   To generalize results from a sample to a population
   To test hypotheses
   To estimate unknown population parameters Common tools:

Probability
  Hypothesis testing (t-test, z-test, chi-square test)
  Confidence intervals
  Regression and correlation

Example:
  Surveying 100 voters to predict an election result
  Testing whether a new teaching method improves student performance
   Estimating the average income of a city using sample data

#2. What is sampling in statistics? Explain the differences between randomand stratified sampling.

Sampling in Statistics

Sampling is the process of selecting a subset (sample) of individuals or observations from a larger group (population) to draw conclusions about the entire population.

Random Sampling
  In random sampling, every member of the population has an equal and independent chance of being selected.

How it works:
  Lottery method
  Random number tables or computer-generated random numbers

Example:
From a class of 100 students, randomly selecting 10 students such that each student has the same chance of selection.


Stratified Sampling
   In stratified sampling, the population is first divided into homogeneous subgroups (strata) based on a characteristic (e.g., age, gender, income), and then random samples are taken from each stratum.

How it works:
  Divide population into strata
  Perform random sampling within each stratum

Example:
A school has:

60% girls

40% boys

To select 10 students:

Select 6 girls randomly

Select 4 boys randomly

#3. Define mean, median, and mode. Explain why these measures of central tendency are important.


Measures of central tendency describe the central or typical value of a dataset. The three most common measures are mean, median, and mode.

mean
  The mean is the arithmetic average of a set of values.

Example:
  For data: 2, 4, 6, 8
  Mean = (2 + 4 + 6 + 8) / 4 = 5

Key Point:
  Sensitive to extreme values (outliers)

median
  The median is the middle value when the data is arranged in ascending or descending order.

Example:
   For data: 3, 5, 7, 9, 11
   Median = 7

For even number of values:
  Data: 4, 6, 8, 10
  Median = (6 + 8) / 2 = 7

Key Point:
  Not affected by extreme values
  Best for skewed distributions   


mode
   The mode is the value that occurs most frequently in a dataset.

  Example:
   For data: 1, 2, 2, 3, 4
   Mode = 2

Key Point:
  Can be one, more than one, or none
  Useful for categorical data

  Zero skewness → Perfectly symmetric distribution (e.g., normal distribution)
  Positive skewness (right-skewed) → Long tail on the right side
  Negative skewness (left-skewed) → Long tail on the left side
  Positive Skewness implies:
  Most data values are concentrated on the lower end
  A few large values (outliers) stretch the distribution to the right
  The typical relationship is:

mathematica
   Mean > Median > Mode

Types of Kurtosis
  Mesokurtic:
  Similar to normal distribution
  Moderate tails (kurtosis ≈ 3)

  Leptokurtic:
  Sharp peak, heavy tails

 Platykurtic:
 Flat peak, light tails

#5.  Implement a Python program to compute the mean, median, and mode ofa given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]


```python
import numpy as np
from scipy import stats

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Calculate the Mean
mean_value = np.mean(numbers)

# Calculate the Median
median_value = np.median(numbers)

# Calculate the Mode
# stats.mode returns a ModeResult object, with mode and count attributes
mode_result = stats.mode(numbers)
mode_value = mode_result.mode # Get the actual mode value, handles scalar or array output

print(f"Given list of numbers: {numbers}")
print(f"\nMean: {mean_value:.2f}")
print(f"Median: {median_value:.2f}")
print(f"Mode: {mode_value}")
```

#6. Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

```python
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Calculate Covariance
covariance_matrix = np.cov(list_x, list_y)
covariance = covariance_matrix[0, 1] # The covariance between list_x and list_y

# Calculate Correlation Coefficient
correlation_coefficient = np.corrcoef(list_x, list_y)[0, 1]

print(f"List X: {list_x}")
print(f"List Y: {list_y}")
print(f"\nCovariance between list_x and list_y: {covariance:.2f}")
print(f"Correlation Coefficient between list_x and list_y: {correlation_coefficient:.4f}")
```

#7. Write a Python script to draw a boxplot for the following numeric list andidentify its outliers. Explain the result:

 data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

A boxplot (or box-and-whisker plot) is a standardized way of displaying the distribution of data based on a five-number summary: the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum.

### How Boxplots Identify Outliers

Outliers are data points that fall outside the overall pattern of the distribution. In a boxplot, outliers are typically identified using the Interquartile Range (IQR).

*   **Interquartile Range (IQR)**: This is the range between the first quartile (Q1) and the third quartile (Q3). IQR = Q3 - Q1.
*   **Upper Bound for Outliers**: Any data point greater than Q3 + 1.5 * IQR is considered an outlier.
*   **Lower Bound for Outliers**: Any data point less than Q1 - 1.5 * IQR is considered an outlier.

Points that fall outside these whiskers are often plotted individually as dots or asterisks, visually marking them as potential outliers.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Create the boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(y=data)
plt.title('Boxplot of Numeric Data')
plt.ylabel('Values')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Identify outliers programmatically
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print(f"\nFirst Quartile (Q1): {Q1}")
print(f"Third Quartile (Q3): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")
print(f"Lower Bound for Outliers: {lower_bound}")
print(f"Upper Bound for Outliers: {upper_bound}")
print(f"Identified Outliers: {outliers}")

if not outliers:
    print("There are no outliers in this dataset based on the 1.5*IQR rule.")
```

#8. You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and  daily sales.
  
  ● Explain how you would use covariance and correlation to explore this
  relationship.
  ● Write Python code to compute the correlation between the two lists:
   advertising_spend = [200, 250, 300, 400, 500]
   daily_sales = [2200, 2450, 2750, 3200, 4000]

To explore the relationship between advertising spend and daily sales, covariance and correlation are two fundamental statistical measures:

### Covariance
Covariance measures the direction of the linear relationship between two variables. In simpler terms, it tells you whether two variables tend to move in the same direction (positive covariance) or in opposite directions (negative covariance).

*   **Positive Covariance**: If advertising spend increases and daily sales also tend to increase, the covariance will be positive. This suggests a direct relationship.
*   **Negative Covariance**: If advertising spend increases and daily sales tend to decrease, the covariance will be negative. This suggests an inverse relationship.
*   **Zero Covariance**: If there's no consistent linear relationship between the two, the covariance will be close to zero.

**Limitation**: The magnitude of covariance depends on the units of the variables, making it difficult to compare the strength of relationships between different pairs of variables. For example, a covariance of 100 between advertising spend (in dollars) and sales (in dollars) doesn't tell us if the relationship is strong or weak without knowing the scale of the variables.

### Correlation
Correlation is a standardized measure that quantifies both the strength and direction of a linear relationship between two variables. Unlike covariance, correlation is unitless and ranges from -1 to +1.

*   **Correlation Coefficient (r)**:
    *   **+1**: Indicates a perfect positive linear relationship (as one variable increases, the other increases proportionally).
    *   **-1**: Indicates a perfect negative linear relationship (as one variable increases, the other decreases proportionally).
    *   **0**: Indicates no linear relationship.
    *   **Values between 0 and 1 (or -1 and 0)**: Indicate the strength of the positive or negative linear relationship. Closer to 1 or -1 means a stronger relationship.

**Why Correlation is Preferred**: For your marketing team, correlation is generally more useful than covariance because its standardized nature allows for easier interpretation and comparison. A high positive correlation (e.g., 0.8 or 0.9) between advertising spend and daily sales would strongly suggest that increasing advertising spend leads to increased sales, providing a clear insight for strategic decisions.

### How to Use Them
1.  **Calculate Covariance**: This is an initial step to see the general direction. If it's positive, you know they move together.
2.  **Calculate Correlation**: This is the crucial step. The correlation coefficient will tell you how strong that linear relationship is. For instance, a strong positive correlation would suggest that your advertising efforts are effective in driving sales.
3.  **Visualize**: Always complement these statistics with scatter plots to visually inspect the relationship. This helps identify non-linear relationships or outliers that might skew your statistical measures.

Let's compute the correlation for your data:

```python
import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Calculate the Pearson correlation coefficient
correlation = np.corrcoef(advertising_spend, daily_sales)[0, 1]

print(f"Advertising Spend: {advertising_spend}")
print(f"Daily Sales: {daily_sales}")
print(f"\nThe correlation coefficient between advertising spend and daily sales is: {correlation:.4f}")
```

#9. Your team has collected customer satisfaction survey data on a scale of1-10 and wants to understand its distribution before launching a new product.

   ● Explain which summary statistics and visualizations (e.g. mean, standard
      deviation, histogram) you’d use.
  ● Write Python code to create a histogram using Matplotlib for the survey data
   survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]


To understand the distribution of your customer satisfaction survey data, you'd typically look at a combination of summary statistics and visualizations:

### Summary Statistics

*   **Mean**: The average survey score. This gives you a central tendency of the data. `(e.g., np.mean(survey_scores))`
*   **Median**: The middle value when the scores are ordered. This is useful for understanding the typical score without being heavily influenced by outliers. `(e.g., np.median(survey_scores))`
*   **Mode**: The most frequently occurring score. This tells you which specific score(s) are most common. `(e.g., stats.mode(survey_scores))`
*   **Standard Deviation**: Measures the spread or dispersion of the scores around the mean. A small standard deviation indicates scores are clustered close to the mean, while a large one suggests wider spread. `(e.g., np.std(survey_scores))`
*   **Min and Max**: The lowest and highest scores recorded. These give you the full range of the data. `(e.g., np.min(survey_scores), np.max(survey_scores))`

### Visualizations

*   **Histogram**: This is an excellent visualization for showing the distribution of numerical data. It displays the frequency of scores within different bins (ranges), allowing you to see the shape of the distribution, identify peaks, and detect skewness or outliers.
*   **Box Plot (Box-and-Whisker Plot)**: This plot visually summarizes the distribution of the data using five key values: minimum, first quartile (25th percentile), median (50th percentile), third quartile (75th percentile), and maximum. It's particularly good for identifying potential outliers and comparing distributions between different groups, if you had them.

Let's generate a histogram for your survey data:

```python
import matplotlib.pyplot as plt
import numpy as np

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(survey_scores, bins=np.arange(3.5, 11.5, 1), edgecolor='black', alpha=0.7)

# Add titles and labels
plt.title('Distribution of Customer Satisfaction Survey Scores')
plt.xlabel('Survey Score (1-10)')
plt.ylabel('Frequency')

# Set x-axis ticks to represent each score clearly
plt.xticks(np.arange(4, 11, 1))

# Add a grid for better readability
plt.grid(axis='y', alpha=0.75)

# Display the histogram
plt.show()
```
